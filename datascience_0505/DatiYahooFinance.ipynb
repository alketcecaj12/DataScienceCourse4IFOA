```python
dati = pd.read_csv("Tesla.csv")
```

```python
dati
```

```python
dati.set_index("Date", inplace = True)
```

```python
dati
```

```python
dati["Return"] = dati["Close"].pct_change() * 100
```

```python
dati.dropna(inplace= True)
```

```python
dati
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (20, 5))
plt.plot(dati["Close"][1200:])
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (20, 4))
plt.plot(dati["Return"])
plt.show()
```
