```python
import pandas as pd
```

```python
dati = pd.read_csv("dati/sales.csv")
```

```python
dati = dati.drop(['Unnamed: 0'], axis = 1)
```

```python
dati.head(2)
```

##### Quanto ha venduto ogni agente ?

```python
vendite_agente = dati.pivot_table(index= ["NomeAgente"], values = ["Vendite"], aggfunc = "sum")
```

```python
vendite_agente
```

#### Quanto sono le vendite totali per ogni mercato/paese?

```python
vendite_paese = dati.pivot_table(index = "Paese", values = "Vendite", aggfunc = "sum")
```

```python
vendite_paese
```

#### Massima vendita (valore) realizata da un agente/paese?

```python
max_vendita_agente = dati.pivot_table(index = ["NomeAgente"], values = ["Vendite"], aggfunc="max")
```

```python
max_vendita_agente
```

```python
max_vendita_paese = dati.pivot_table(index = ["Paese"], values = ["Vendite"], aggfunc="max")
```

```python
max_vendita_paese
```

```python
vendite_ag_p = dati.pivot_table(index = ["NomeAgente", "Paese"], values = "Vendite")
```

```python
vendite_ag_p
```

#### Qual è la media, mediana e il min vendite per paese ?

```python
dati_agg_stats = dati.pivot_table(index = ["Paese"], values = ["Vendite"], aggfunc={"mean", "median", "min"})
```

```python
dati_agg_stats
```

```python
dati_agg_stats_ag = dati.pivot_table(index = ["NomeAgente"], values = ["Vendite"], aggfunc={"mean", "median", "min"})
```
