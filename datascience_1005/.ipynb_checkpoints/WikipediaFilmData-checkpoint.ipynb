```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
url = "https://en.wikipedia.org/wiki/List_of_action_films_of_the_2010s"
```

```python
dati_g = pd.read_html(url)
```

```python
len(dati_g)
```

```python
dati_tot = pd.DataFrame({"Title": [], "Director":[], "Cast":[], "Country":[], "Subgenere/notes":[], "Unnamed: 5":[], "Unnamed: 6": []})
```

```python
index = 1
for i in dati_g: 
    print(dati_g[index])
    
    dati_tot = dati_tot.append(dati_g[index][1:])
    
    index = index + 1
    if index == 11: 
        break
```

```python
dati_tot
```

```python
dati_tot = dati_tot.drop(["Unnamed: 5", "Unnamed: 6"], axis = 1)
dati_tot
```

```python
for i, row in dati_tot.iterrows(): 
    print(row["Title"], row["Director"], row["Cast"])
```

```python
group_by_country = dati_tot.groupby("Country")
```

```python
countries = {}
for index, frame in group_by_country: 
    #print(index, len(frame))
    countries[index] = len(frame)
```

```python
countries
```

```python
df_countries_prod = pd.DataFrame(list(countries.items()), columns = ["Paese-CP", "NumeroFilm"])
```

```python
df_sorted = df_countries_prod.sort_values("NumeroFilm", ascending = False)
df_sorted
```

```python
df_sorted = df_sorted[:20]
```

```python
plt.figure(figsize = (15, 5))
plt.bar(df_sorted["Paese-CP"], df_sorted["NumeroFilm"])
plt.xticks(rotation = 45)
plt.show()
```
