```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
c15 = pd.read_csv("dati/diffusionecognomi2015.csv", sep = ";")
c19 = pd.read_csv("dati/diffusionecognomi2019.csv")
```

```python
c19
```

```python
plt.figure(figsize = (6,6))
plt.barh(c19["COGNOME"][:20], c19["DIFFUSIONE (FREQUENZA)"][:20])
plt.show()
```

```python
li_cognomi = []
for i, row in c15.iterrows(): 
    ci = row["COGNOME"]
    if " " in ci: 
        ci = ci.replace(" ", "")
        
    fi = row["DIFFUSIONE"]
    
    ci = (ci+" ") * fi
    li_cognomi.append(ci)
```

```python
len(li_cognomi)
    
```

```python
l = ["ciao", "hello", "hola"]
stringa_tot = " ".join(l)
stringa_tot
```

```python
contenitore = " "
text_wordcloud = contenitore.join(li_cognomi)
text_wordcloud
```

```python
wordcloud_grahics = WordCloud(collocations = False).generate(text_wordcloud)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(wordcloud_grahics, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
