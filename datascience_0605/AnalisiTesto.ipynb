```python
import bs4 
import requests
```

```python
url = "https://it.wikipedia.org/wiki/Italia"
```

```python
page = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(page.content, "html.parser")
```

```python
lista_tag = soup.find_all("p")
```

```python
lista_link = soup.find_all("a")
lista_link
```

```python
type(lista_tag)
```

```python
lista_tag[5].get_text()
```

```python
index = 1
lista_testo = []
for i in lista_tag: 
    print(lista_tag[index].get_text())
    lista_testo.append(lista_tag[index].get_text())
    index += 1
    
```

```python
stringa = ""
articolo_italia = stringa.join(lista_testo)
```

```python
type(articolo_italia)
```

```python
articolo_italia
```

```python
articolo_italia = articolo_italia.replace("di", "")
articolo_italia = articolo_italia.replace("e", "")
articolo_italia = articolo_italia.replace("il", "")
articolo_italia = articolo_italia.replace("la", "")
articolo_italia = articolo_italia.replace("le", "")
articolo_italia = articolo_italia.replace("che", "")
articolo_italia = articolo_italia.replace("un", "")
articolo_italia = articolo_italia.replace("più", "")
articolo_italia = articolo_italia.replace("della", "")
articolo_italia = articolo_italia.replace(".", "")
articolo_italia = articolo_italia.replace(",", "")
articolo_italia = articolo_italia.replace("\n", "")
articolo_italia = articolo_italia.replace("\"", "")  
articolo_italia = articolo_italia.replace("\xa0", "")
```

```python
articolo_italia
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud( background_color = "white").generate(articolo_italia)
plt.figure(figsize = (10, 8))
plt.imshow(wc)
plt.axis("off")
```
