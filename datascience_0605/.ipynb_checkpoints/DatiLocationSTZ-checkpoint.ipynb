```python
import pandas as pd 
import folium 
```

```python
dati = pd.read_csv("QARIA Stazioni - Foglio1.csv")
```

```python
dati
```

```python
m_lon = dati["LON_GEO"].mean()
m_lat = dati["LAT_GEO"].mean()
```

```python
print(m_lon, m_lat)
```

```python
mappa = folium.Map(location = [m_lat, m_lon], zoom_start = 8)
```

```python
mappa
```

```python
for i in range(0, len(dati)): 
    folium.Marker([dati.iloc[i]["LAT_GEO"], dati.iloc[i]["LON_GEO"]], popup = dati.iloc[i]["Stazione"]).add_to(mappa)
```

```python
mappa
```

```python
mappa.save("MappaStzQA.html")
```
