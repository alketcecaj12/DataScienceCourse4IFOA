```python
import pandas as pd 
import yfinance as yf 
import matplotlib.pyplot as plt 
```

```python
t = yf.Ticker("MSFT")
```

```python
dati = t.history(period = "10Y")
```

```python
dati
```

```python
dati.to_csv("Azioni_MSFT.csv")
```

```python
dati["Return"] = dati["Close"].pct_change() * 100
```

```python
dati.dropna(inplace = True)
```

```python
plt.figure(figsize = (20, 5))
plt.plot(dati["Close"][2200:])
plt.show()
```

```python
dati
```

```python
gg_positivi = dati[dati["Return"] >  0]
gg_negativi = dati[dati["Return"] <= 0]
```

```python
# VolatilitÃ 
dati["Return"].std()
```

```python
dati["Return"].mean()
```

```python
plt.figure(figsize = (20, 4))
plt.plot(dati["Return"])
plt.show()
```
