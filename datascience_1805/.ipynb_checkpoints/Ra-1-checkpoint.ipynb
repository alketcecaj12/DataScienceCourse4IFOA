```python
import pandas as pd 
```

```python
rating = pd.read_csv("dati/ratings.csv")
movie_names = pd.read_csv("dati/movie_names.csv")
```

```python
rating.drop(["Unnamed: 0"], axis = 1, inplace = True)
movie_names.drop(["Unnamed: 0"], axis = 1, inplace = True)
```

```python
print(rating.head(2))
print(movie_names.head(2))
```

```python
combined_movies_data = pd.merge(rating, movie_names, on = "item_id")
```

```python
combined_movies_data
```

```python
combined_movies_data.groupby("item_id")["rating"].count().sort_values(ascending=False).head(10)
```

```python
film_x = combined_movies_data["item_id"] == 50
combined_movies_data[film_x]['movie title'].unique()
```

##### Costruisco la Utility Matrix

```python
rating_crosstab = combined_movies_data.pivot_table(index= "user_id", values= "rating", columns = "movie title", fill_value=0)
```

```python
rating_crosstab
```

#### Calcolo la trasposta della matrice utilità

```python
X = rating_crosstab.T
X
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
SVD = TruncatedSVD(n_components=10, random_state = 1)
```

```python
resultant_matrix = SVD.fit_transform(X)
```

```python
type(resultant_matrix)
```

```python
resultant_matrix
```

```python
import numpy as np
corr_mat = np.corrcoef(resultant_matrix)
```

```python
corr_mat
```
