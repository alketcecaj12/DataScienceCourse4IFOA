#### Aggregazione dati

```python
film_agg_per_gen = dati.groupby("genre")
```

```python
type(film_agg_per_gen)
```

```python
for i, frame in film_agg_per_gen: 
    print(i, len(frame))
```

```python
for i, frame in film_agg_per_gen: 
    print(i, frame["star_rating"].mean())
```

```python
for i, frame in film_agg_per_gen: 
    print(i, frame)
```

```python
film_agg_by_cr = dati.groupby("content_rating")
```

```python
for i, frame in film_agg_by_cr: 
    print(i, frame)
```
