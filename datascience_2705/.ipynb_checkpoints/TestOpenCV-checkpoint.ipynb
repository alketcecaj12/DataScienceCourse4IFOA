```python
altezza = img.shape[0]
largezza = img.shape[1]
print(altezza, largezza)
```

```python
new_alt = int(img.shape[0] * 0.5)
new_larg = int(img.shape[1] * 0.5)
```

```python
new_dim = ( new_larg, new_alt)
new_dim
```
