```python
while(video.isOpened()): 
    ret, frame = video.read()
    if ret == True:
        cv2.imshow("Video Chaplin!", frame)
        cv2.waitKey(200)
        
    else: 
        break
        
cv2.destroyAllWindows()        
        
```

#### Accedere alla webcam del PC. 

```python
cap = cv2.VideoCapture(0)
```

```python
type(cap)
```

```python
while(cap.isOpened):
    ret,frame = cap.read()
    cv2.imshow("Webcam ", frame)
    
    if cv2.waitKey(250) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
