```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
import os
```

```python
cartella = "Tesseract2Scontrini"
```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:/Program Files/Tesseract-OCR/tesseract.exe"
```

```python
lista_scontrini = os.listdir(cartella)
```

```python
scontrini = {}
for i in lista_scontrini: 
    print(i)
    testo_img = pytesseract.image_to_string(cartella+"/"+i)
    print(testo_img)
    testo_img= testo_img.split("\n")
    scontrini[i] = testo_img
    print("------------------------------------------------------------")
```

```python
scontrini
```

```python
with open("DatiScontrini.csv", "w") as file: 
    for key, values in scontrini.items(): 
        file.write(str(key)+", "+str(values)+"\n")
```
